```python
# vytvorenie triedy perceptron
class Perceptron():   
    def __init__(self, rychlost_ucenia=0.01, pocet_iteracii=15): # konštruktor - inicializuje atribúty triedy (rýchlosť učenia, počet iterácií)
        self.eta = rychlost_ucenia # priradenie do premennej - prístup k atribútom
        self.n_iter = pocet_iteracii # priradenie do premennej - prístup k atribútom
    
    
    def fit(self, X,y): # metóda procesu učenia 
        self.vahy = np.zeros(1+ X.shape[1]) # šírka poľa váh 
        self.errors_=[] # vytvorenie poľa erorov
        
        # vytvorenie premenných typu pole na vitualizáciu váh
        self.vahy_to_visualize_1=[] 
        self.vahy_to_visualize_2=[]
        self.vahy_to_visualize_3=[]
       
        # cyklus na naplnenie
        for i in range(self.n_iter): 
            errors = 0
            for matica,vystup in zip(X,y):
                update = self.eta * (vystup-self.predikcia(matica))
                self.vahy[1:] += update*matica
                self.vahy[0]+= update
                
                print(self.vahy[1:]) # výpis váh
                
                errors += int(update !=0.0)
                 # naplnenie premenných
            self.errors_.append(errors)   
            self.vahy_to_visualize_1.append(self.vahy[0])
            self.vahy_to_visualize_2.append(self.vahy[1])
            self.vahy_to_visualize_3.append(self.vahy[2])
        return self  

    def net_input(self, X): # výpočet čistého vstupu
        return np.dot(X, self.vahy[1:])+ self.vahy[0]

    def predikcia(self, X): # návrat triedy po kroku
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
import matplotlib.pyplot as plt # importovanie knižnice pre prácu s grafmi
```

```python
X = df_new.iloc[0:3000, [1,2]].values # výber stĺpca s indexom 0 a 2 z údajov premennej df_new
y = df_new.iloc[0:3000, 4].values # výber stĺpca s indexom 4 z údajov premennej df
y = np.where(y == 'Richmond', -1, 1) # vytvorenie matice z  -1 a 1 podľa zadaného názvu
print(y) # výpis matice
```

```python
moj_perceptron_Richmond_teplota = Perceptron() # vytvorenie premennej a nadobudnutie atribútov triedy perceptron
moj_perceptron_Richmond_teplota.fit(X,y) # výpis premenej procesesom učenia X,y
```

```python
plt.plot(moj_perceptron_Richmond_teplota.errors_) # vykreslenie grafu chyby pre mesto Richmond
```
